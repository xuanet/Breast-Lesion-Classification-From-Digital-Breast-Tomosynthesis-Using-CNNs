## Transforming Images with Physical Layers

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

lpf = np.ones(shape=(4, 4))

"""number of detectors"""
def low_res(img):
    img = sp.signal.convolve2d(img, lpf, mode='same') / lpf.size
    return img


"""voltage (intensity), detector sensitivity"""
def low_contrast(image, degree):
    # Apply gamma correction
    d = 1 + np.exp(-degree*(image-0.5))
    return 1 / d
```

## Training set

```python
training_array_raw = np.load("training_array_RAW.npy")
```

```python
plt.imshow(training_array_raw[1, :, :], cmap='gray', vmin = 0, vmax = 1)
plt.show()
```

### Physical layer 1: low res

```python
training_array_LOW_RES = np.array([low_res(img) for img in training_array_raw])
```

```python
plt.imshow(training_array_LOW_RES[1, :, :], cmap='gray', vmin = 0, vmax = 1)
plt.show()
```

```python
np.save('training_array_LOW_RES', training_array_LOW_RES)
```

### Physical layer 2: low contrast

```python
training_array_LOW_CONTRAST = np.array([low_contrast(img, 3) for img in training_array_raw])
```

```python
plt.imshow(training_array_LOW_CONTRAST[1, :, :], cmap='gray', vmin = 0, vmax = 1)
plt.show()
```

```python
np.save('training_array_LOW_CONTRAST', training_array_LOW_CONTRAST)
```

### Physical layer 3: combined

```python
training_array_COMBINED = np.array([low_contrast(img, 3) for img in training_array_LOW_RES])
```

```python
plt.imshow(training_array_COMBINED[1, :, :], cmap='gray', vmin = 0, vmax = 1)
plt.show()
```

```python
np.save('training_array_COMBINED', training_array_COMBINED)
```
